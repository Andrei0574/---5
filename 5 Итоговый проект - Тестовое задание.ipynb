2. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен DAU 

```python
import pandas as pd

audience_df = pd.read_excel(
    'Данные для тестового задания.xlsx',
    sheet_name='Данные об аудитории'
)

# Количество уникальных пользователей по дням
dau_daily = audience_df.groupby('date')['user_id'].nunique()

# Средний DAU за месяц
dau = dau_daily.mean()

print('DAU:', round(dau, 2))
```

3. Используя вкладку "Данные об аудитории", посчитайте, чему будет равен retention первого дня у пользователей, пришедших в продукт 1 ноября

```python
import pandas as pd

audience_df = pd.read_excel(
    'Данные для тестового задания.xlsx',
    sheet_name='Данные об аудитории'
)

audience_df['date'] = pd.to_datetime(audience_df['date'])

# Пользователи, пришедшие 1 ноября
users_1_nov = set(
    audience_df[audience_df['date'] == '2023-11-01']['user_id']
)

# Пользователи, посетившие продукт 2 ноября
users_2_nov = set(
    audience_df[audience_df['date'] == '2023-11-02']['user_id']
)

# Вернувшиеся пользователи
returned_users = users_1_nov & users_2_nov

retention_d1 = len(returned_users) / len(users_1_nov)

print('Пользователей 1 ноября:', len(users_1_nov))
print('Вернулось 2 ноября:', len(returned_users))
print('Retention D1:', round(retention_d1 * 100, 2), '%')
```

5. Посчитайте пользовательскую конверсию в просмотр объявления за ноябрь? (в пользователях) 

```python
user_adverts = audience_df.groupby('user_id')['view_adverts'].sum().reset_index()

total_users = user_adverts['user_id'].nunique()
users_with_adverts = user_adverts[user_adverts['view_adverts'] > 0]['user_id'].nunique()

conversion = users_with_adverts / total_users * 100

print('Конверсия в просмотр объявления:', round(conversion, 2), '%')
```

6. Используя информацию из вкладки "Данные об аудитории", посчитайте среднее количество просмотренных объявлений на пользователя в ноябре

```python
total_adverts = audience_df['view_adverts'].sum()

# Количество уникальных пользователей
total_users = audience_df['user_id'].nunique()

# Среднее количество объявлений на пользователя
avg_adverts = total_adverts / total_users

print('Всего просмотров объявлений:', total_adverts)
print('Количество пользователей:', total_users)
print('Среднее количество просмотров на пользователя:', round(avg_adverts, 2))
```

8. Посмотрите на результаты тестов и интерпретируйте их. Напишите значения p-value, которые вы получили.

```python
import pandas as pd

ab_df = pd.read_excel(
    'Данные для тестового задания.xlsx',
    sheet_name='Данные АБ тестов'
)

print(ab_df.head())
print(ab_df.info())
```

```python
print(ab_df['experiment_num'].unique())
```

```python
from scipy.stats import ttest_ind

for exp in sorted(ab_df['experiment_num'].unique()):

    exp_data = ab_df[ab_df['experiment_num'] == exp]

    control = exp_data[
        exp_data['experiment_group'] == 'control'
    ]['revenue']

    test = exp_data[
        exp_data['experiment_group'] == 'test'
    ]['revenue']

    stat, p_value = ttest_ind(
        control,
        test,
        equal_var=False
    )

    print(f'Эксперимент {exp}')
    print(f'p-value = {p_value:.6f}')
    print(f'ARPU Control = {control.mean():.2f}')
    print(f'ARPU Test = {test.mean():.2f}')
    print('-' * 40)
```

9. По датасету с листерами посчитайте средний доход на пользователя 

```python
import pandas as pd

file_path = 'Данные для тестового задания.xlsx'

listers_df = pd.read_excel(
    file_path,
    sheet_name='Листеры'
)

print(listers_df.head())
print(listers_df.columns)

user_revenue = (
    listers_df
    .groupby('user_id', as_index=False)
    .agg({'revenue': 'sum'})
)

avg_revenue_per_user = user_revenue['revenue'].mean()

print('Количество пользователей:', user_revenue['user_id'].nunique())
print('Средний доход на пользователя:', round(avg_revenue_per_user, 2))
```
